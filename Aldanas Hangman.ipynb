# Hangman game

This was a challenge given to me in the Primer part of a Data Sience course that I took a WBS CODING SCHOOL. The school gave some instructions to develop the exercise. I followed some of them. The graphic of the hangman used here is the one given by the school.

Hangman is a guessing game where the player tries to guess a hidden word by suggesting letters.

To start, the player is given a word where all letters are replaced with underscores (`_`) so the only information available is the number of letters in the answer.

Each round the player guesses a letter and if the letter guesed is in the answer, its position (or positions) in the word is revealed. Once they successfully fill in the hidden word with its correct letters, the player wins!

If the letter guessed is not in the answer, the player accumulates mistakes. If they make it to six mistakes without finding the answer, the player loses!

The name of the game comes from a drawing of a hanged stick figure. After each mistake, a new element of the hanged figure is drawn.

If the player doesn't guess the word before the last element of the figure is drawn, the player is hanged!

### **Challenge's solution**

```python
# Hangman Project - Clara Aldana's code

import random
list_words = ["RABBIT", "WHALE", "RHINOCEROS", "TUCAN", "KINGFISHER", "HEDGEHOG", "LOBSTER", "ELEPHANT", "TIGER", "CARIBOU", "PENGUIN", "GECKO"]
answer_word = random.choice(list_words)
aw_length = len(answer_word)  # The length of the word

# We define the hidden word accordinly to the length of the answer word
hidden_word = str("")
for m in range(0,2*aw_length):
  if m % 2 == 0:  # m even
    hidden_word = hidden_word + "_"
  else: # m odd
    hidden_word = hidden_word + " "

###############################################################
hangman={0:'''
        ____________
         |



         ''',
        1:'''
        ____________
         |
         O


         ''',
        2:'''
        ____________
         |
         O
        /

        ''',
        3:'''
        ____________
         |
         O
        / \\

        ''',
        4:'''
        ____________
         |
         O
        / \\
         |
         ''',
        5:'''
        ____________
         |
         O
        / \\
         |
        /''',
        6:'''
        ____________
         |
         O
        / \\
         |
        / \\ '''}

###################################################
print(f"Welcome! This is a Hangman game.\n You should guess a word from a hidden list of words.\n In this game, the words are names of animals.\n You can make {6} mistakes before losing.\n Good luck!")

remaining_mistakes = 6

while remaining_mistakes > 0:
  print("============================================== \n")
  print(f"The hidden word is: {hidden_word} \n")
  picked_letter = str(input("Pick a letter: "))
  picked_letter = picked_letter.upper()

  for letter in answer_word:
  #  index_count = 0 # for repeated letters we do not need it
    if letter == picked_letter: # This line checks if the letter is in the answer word
  #    index_count = answer_word.count(picked_letter) # gives the number of appearances of letter in the word we do not need it
      for n in range(0,aw_length):  # we run n through all the word to be able to change all appearances of the letter
        if letter == answer_word[n]:
          temp_index = n  # gives the position of the nth appearance of the letter
          stemp_index = 2 * temp_index
          hidden_word = hidden_word[:stemp_index] + picked_letter + hidden_word[stemp_index + 1:]

################
  if picked_letter in answer_word:
    print(f"CORRECT! The word contains the letter {picked_letter}!")
    print(hangman[6-remaining_mistakes])
  else:
    remaining_mistakes -= 1
    print(f"WRONG! Number of mistakes left: {remaining_mistakes}")
    print(hangman[6-remaining_mistakes])

  letters_left = hidden_word.count("_")
  if letters_left == 0:
    if remaining_mistakes == 0:
      print(f"The hidden word was: {answer_word}")
      print("HANGED!")
    else:
      print(hidden_word)
      print("CONGRATULATIONS!")
      break

if remaining_mistakes == 0:
      print(f"The hidden word was: {answer_word}")
      print("HANGED!")
```
